```python
import matplotlib.pyplot as plt
```

```python
def find_base(x):
    while x % 2 == 0:
        x //= 2
    return x

x = []
y = []

for i in list(range(1, 200, 2)):
    j = find_base(3 * i + 1)
    
    x.append(i)
    y.append(j)

plt.figure(figsize=(8, 6), dpi=80)
plt.grid()

plt.plot(x, y, '.-b')
plt.plot(x, [3 * i / 2 + 1 / 2 for i in x], '.-r')
plt.plot(x, [3 * i / 4 + 1 / 4 for i in x], '.-g')
plt.plot(x, [3 * i / 8 + 1 / 8 for i in x], '.-y')
```
